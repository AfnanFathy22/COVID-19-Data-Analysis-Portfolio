```python
num_cols=data.select_dtypes(include='number').columns
cat_cols=data.select_dtypes(include='object').columns
```

```python
data[num_cols].columns
```

```python
data[cat_cols].columns
```

```python
for col in num_cols:
    if col in ["NewCases", "NewDeaths", "NewRecovered", "Serious,Critical"]:
        data[col] = data[col].fillna(0)
    else:
        data[col] = data[col].fillna(data[col].median())
data[cat_cols] = data[cat_cols].fillna("Unknown")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.drop(
    ["NewCases","NewDeaths","NewRecovered"],
    axis=1,
    inplace=True
)
```

```python
data.columns = (
    data.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
    .str.replace(",", "")
)
```

```python
data["population"] = data["population"].astype(int)
```

```python
data = pd.get_dummies(
    data,
    columns=["continent","who_region"],
    drop_first=True
)
```
