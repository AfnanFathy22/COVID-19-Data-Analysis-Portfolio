```python
data.columns = (
    data.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
)
```

```python
data["date"] = pd.to_datetime(data["date"])
```

```python
data
```

```python
num_cols = [
    "confirmed","deaths","recovered","active",
    "new_cases","new_deaths","new_recovered"
]

for col in num_cols:
    data[col] = pd.to_numeric(data[col], errors="coerce")

```

```python
data.columns
```

```python
data['Death Rate'] = np.where(data['confirmed']>0, data['deaths']/data['confirmed'], 0)
data['Recovery Rate'] = np.where(data['confirmed']>0, data['recovered']/data['confirmed'], 0)
data['Active Rate'] = np.where(data['confirmed']>0, data['active']/data['confirmed'], 0)
```

```python
data.info()
```

```python
data["year"] = data["date"].dt.year
data["month"] = data["date"].dt.month
data["day"] = data["date"].dt.day
data["dayofweek"] = data["date"].dt.dayofweek
```

```python
data["cases_7d_avg"] = (
    data.groupby("country_region")["new_cases"]
    .transform(lambda x: x.rolling(7).mean())
)

data["deaths_7d_avg"] = (
    data.groupby("country_region")["new_deaths"]
    .transform(lambda x: x.rolling(7).mean())
)
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Country_Code'] = le.fit_transform(data['country_region'])

le_region = LabelEncoder()
data['Region_Code'] = le.fit_transform(data['who_region'])

data['iso_alpha'] = le.fit_transform(data['iso_alpha'])
```

```python
data = data.sort_values(['country_region', 'date']).reset_index(drop=True)
data.drop(columns=['country_region','who_region'] , inplace=True)
```

```python
data
```

```python
data.fillna(0, inplace=True)
data[['cases_7d_avg', 'deaths_7d_avg']] = data[['cases_7d_avg', 'deaths_7d_avg']].interpolate()
```
